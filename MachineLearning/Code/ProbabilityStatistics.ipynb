```python
import numpy as np
import pandas as pd

df = pd.read_csv('../data/house_prices.csv', encoding='cp949')
df
```

### 방의 개수가 7개 초과일 확률

```python
rm = df[df['RM'] > 7]
rm
```

```python
n_rm = len(rm)
print(n_rm)
```

```python
n = len(df)
print(n)
```

```python
p = n_rm / n
print(p)
```

### 조건부 확률

```python
ptratio = df[df['PTRATIO'] < 15]
ptratio.head(10)
```

```python
p_ptratio = len(ptratio) / len(df)
print(p_ptratio)
```

```python
ptratio_rm = df[(df['RM'] > 7) & (df['PTRATIO'] < 15)]
ptratio_rm.head(10)
```

```python
p_ptratio_rm = len(ptratio_rm) / len(df)
print(p_ptratio_rm)
```

```python
p = p_ptratio_rm / p_ptratio
print(p)
```

### 평균, 분산, 표준편차

평균

분산

```python
# n-1 로 나눔
df['RM'].std()
```
